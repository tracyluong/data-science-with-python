```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="bus_map")
location_sf = geolocator.geocode("San Francisco, CA, USA")
print((location_sf.latitude, location_sf.longitude))
```

```python
# create San Francisco map
sanfran_map = folium.Map(location=[location_sf.latitude, location_sf.longitude], zoom_start=12)
```

```python
# instantiate a feature group for the potential bus stops in the dataframe
bus_map = folium.map.FeatureGroup()

# loop through all rows and add each to the bus map feature group
for lat, lng, in zip(df_intersection.latitude, df_intersection.longitude):
    bus_map.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add potential bus stops to map
sanfran_map.add_child(bus_map)
```
